Data Cleaning



```python
# checking null values
df.isnull().sum().sort_values(ascending=False)
     
```

1.Released_Year Column


```python
# Counting values of 'Released_Year' Column
df['Released_Year'].value_counts()
```

```python
# finding row where Released_Year = PG
x = df['Released_Year'].loc[df['Released_Year']=='PG']
x
```

```python
# Dropping the row where Released_Year = PG by its index value
df = df.drop(labels=966, axis=0)

# showing unique values
df['Released_Year'].unique()
```

```python
# Conversion of data type from Object to int
df['Released_Year'] = df['Released_Year'].astype(int)

# Checking Data Type of Released_Year Column
print(df['Released_Year'].dtypes)
```

2.Runtime Column


```python
df['Runtime'].unique()
```

```python
# Checking how many values in 'Runtime' column is digits
df['Runtime'].str.isdigit().sum() 
```

```python
df['Runtime'] = df['Runtime'].str.split(' ').str.get(0).astype(int)
```

```python
# Checking unique values after changing 'Runtime' Column Data.
df['Runtime'].unique()
```

3.Gross Column


```python
# Checking unique Values
df['Gross'].unique()
     
```

```python
# Changing the 'Gross' Data by using the str.replace() function

# String replace() in Python returns a copy of the string where occurrences of a substring are replaced with another substring. 
df['Gross'] = df['Gross'].str.replace(',','')

# Conversion of data type from Object to Float by using astype() function
df['Gross'] = df['Gross'].astype(float)

# showing unique values
print(df['Gross'].unique())

# showing data type of the column
df['Gross'].dtype
     
```

```python
# checking null values
df['Gross'].isna().sum()
     
```

There are 169 null values are present, which is much of 999 values so we will not drop these rows we will fill these values. For treating null values first we will check outliers.

```python
# Checking Outliers
df['Gross'].plot(kind='box')
```

Hence, Outliers found so we will fill these values with 0

Filing null values of the 'Gross' Column

```python
# In this column we fill the Nan values with 0 by using numpy
df['Gross'] = df['Gross'].replace(np.nan,0)

# checking if there is any null values left in this Column
df['Gross'].isna().any()
```

4.Meta_score


```python
# showing unique values in 'Meta_score' Column
df['Meta_score'].unique()
```

```python
# checking for outliers
df['Meta_score'].plot(kind='box')
     
```

Filling null values in 'Meta_score' column

It is seen that there are lower bound outliers found so we fill null values with median.

```python
# Finding median of 'Meta_score'
df['Meta_score'].median()
```

```python
# filling null values with median value of 'Meta_score'
df['Meta_score'].fillna(df['Meta_score'].median(),inplace=True)

# checking for null values
df['Meta_score'].isna().any()
```

```python
# checking for unique values
df['Meta_score'].unique()
```

# Information about the each certificates

```python

# Let's bring some categories together by using .loc() method
df.loc[df['Certificate']=='U/A','Certificate'] = 'UA'
df.loc[df['Certificate']=='G','Certificate'] = 'U'
df.loc[df['Certificate']=='Approved', 'Certificate'] = 'Passed'
df.loc[df['Certificate']=='TV-PG', 'Certificate'] = 'PG'
df.loc[df['Certificate']=='GP', 'Certificate'] = 'PG'
df.loc[df['Certificate']=='TV-14', 'Certificate'] = 'PG-13'
df.loc[df['Certificate']=='16', 'Certificate'] = 'R'
df.loc[df['Certificate']=='TV-MA', 'Certificate'] = 'R'

# assuming missing certicate data means movies are unrated
# filling null values with Unrated one
df.loc[df['Certificate'].isna(), 'Certificate'] = 'Unrated'

# Counting values of Certificate Column
df['Certificate'].value_counts()
     
```

```python
# Plotting Percentage Distribution of movies on the basis of 'Certificate'
plt.figure(figsize=(12,10))
df['Certificate'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.title('Certificate Count Percentage Distribution')
plt.legend(df['Certificate'].value_counts().index)
plt.show()
```

In this plot, The most movies which are of certificate 'U' are rated in top 1000 of imdb because the U certificate movies are friendly to all categories of the people(adults,old,children) etc.
